### Crosstab Analyses

```python
copiedDf.dtypes
```

```python
crosstab_columns = ["Branch","City","CustomerType","Gender","ProductLine","PaymentMethod","DayTime","DayName","MonthName"]
```

```python
for col1 in crosstab_columns:
    for col2 in crosstab_columns:
        if(col1 != col2):
            print(pd.crosstab(copiedDf[col1],copiedDf[col2]))
            print("*****************************************")

        else:
            print("*****************************************")
            continue
```

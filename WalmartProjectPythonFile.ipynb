```python
## I've used ward method and euclidean metric to minimize the difference of variances of clusters
```

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")


linkage_data = linkage(segmentData, method='ward', metric='euclidean')
dendrogram(linkage_data)

plt.show()
```

```python
# n_clusters = 5 gives the best result for both hiearchical and k-means clustering so I've set n_clusters to 5
hierarchical_cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
labels = hierarchical_cluster.fit_predict(segmentData)
segmentData["Labels"] = labels
```

 ## MULTIPLE REGRESSION FOR TOTAL VALUES

```python
newDataset.head()
```

```python
newDataset.columns
```

```python
#Importing Libraries
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
X = newDataset[['UnitPrice','Quantity','Tax']]
y = newDataset['Total']
```

```python
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
regressionModel = LinearRegression()
```

```python
#Implementing the model
regressionModel.fit(X_train, y_train)
```

```python
#Intercept and Coefficient
print("Intercept: ", regressionModel.intercept_)
print("Coefficients:")
list(zip(X, regressionModel.coef_))
```

```python
#Predicting the Test and Train results 
y_pred= regressionModel.predict(X_test)  
x_pred= regressionModel.predict(X_train) 
```

```python
print("Prediction for test set: {}".format(y_pred))
```

```python
#Actual value and the predicted value
reg_model_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred})
reg_model_diff
```

```python
#Model Evaluation
from sklearn import metrics
meanAbErr = metrics.mean_absolute_error(y_test, y_pred)
meanSqErr = metrics.mean_squared_error(y_test, y_pred)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print('R squared: {:.2f}'.format(regressionModel.score(X,y)*100))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```
